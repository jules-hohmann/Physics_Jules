# Simulation

```python
import numpy as np
```

## Parameters

```python
density_water = 1000  # kg/m^3
density_object = 600  # kg/m^3
object_volume = 0.1  # m^3
g = 9.81  # m/s^2
time_step = 0.01  # s
time_max = 10  # s
```

```python
  # s
depth = 0  # m
velocity = 0  # m/s
acceleration = 0  # m/s^2
buoyancy =  density_water * g * object_volume # N
mass = object_volume * density_object  # N
gravitationalforce = mass * g
```

```python
# lists for plotting later
time = np.arange(0, time_max, time_step)
positionlist = np.zeros_like(time)
velocitylist = np.zeros_like(time)
accelerationlist = np.zeros_like(time)
```

```python
# simulation loop
for i in range(1, len(time)):
   # calculate the force on the object
    forcetotal = buoyancy - gravitationalforce
    # calculate acceleration
    acceleration = forcetotal / mass
    accelerationlist[i] = acceleration
    # calculate velocity
    velocity = acceleration * time[i]
    velocitylist[i] =velocity
    # calculate position
    position = 0.5 * acceleration * np.power(time[i],2) + velocity * time[i]
    positionlist[i] = position
    pass
```

# Plot the results

```python
import matplotlib.pyplot as plt
```

```python

plt.plot(time, positionlist, label="Position")
plt.plot(time, velocitylist, label="Velocity")
plt.plot(time, accelerationlist, label="Acceleration")
plt.xlabel("Time (s)")
plt.ylabel("Position (m), Velocity (m/s), Acceleration (m/s^2)")
plt.legend()
plt.show()
```
